```python
from pathlib import Path
import pandas as pd
import tarfile
import urllib.request

def load_housing_data():
    tarball_path = Path("datasets/housing.tgz")
    if not tarball_path.is_file():
        Path("datasets").mkdir(parents=True, exist_ok=True)
        url = "https://github.com/ageron/data/raw/main/housing.tgz"
        urllib.request.urlretrieve(url, tarball_path)
    with tarfile.open(tarball_path) as housing_tarball:
            housing_tarball.extractall(path="datasets")
    return pd.read_csv(Path("datasets/housing/housing.csv"))

housing = load_housing_data()

```

```python
import numpy as np
import pandas as pd
from zlib import crc32

def is_id_in_test_set(identifier, test_ratio):
    return crc32(np.int64(identifier)) < test_ratio * 2**32

def stratified_hash_split(data, test_ratio, id_column, strat_column):

    train_parts = []
    test_parts = []

    for category in data[strat_column].unique():
        subset = data[data[strat_column] == category]
        ids = subset[id_column]
        in_test = ids.apply(lambda x: is_id_in_test_set(x, test_ratio))
        test_parts.append(subset[in_test])
        train_parts.append(subset[~in_test])

    return pd.concat(train_parts), pd.concat(test_parts)
```

```python
housing["id"] = housing["longitude"] * 1000 + housing["latitude"]
housing["income_cat"] = pd.cut(housing["median_income"],
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])

train, test = stratified_hash_split(housing, 0.2, id_column="id", strat_column="income_cat")
```
